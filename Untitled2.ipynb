```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Embedding, SimpleRNN, LSTM
```

```python
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
import re
```

```python
with open('lastofus.txt', encoding='euckr') as f:
  line = f.readlines()

for i in line:
  print(i)
```

```python
text = ''

array = []

with open('lastofus.txt', encoding='euckr') as f:
  line = f.readlines()
  text = text.join([i for i in line if re.match(r"[A-Z].*:", i)])

text
```

```python
print(text)
```

```python
arr = []
for i in text.split('\n'):
  arr.append(i)
```

```python
tokenizer = Tokenizer(lower=False, filters= '..?::\'\'-!%\n')
tokenizer.fit_on_texts(arr)
```

```python
tokenizer.word_index
```

```python
len(tokenizer.word_index)
```

```python
seq = tokenizer.texts_to_sequences(arr)
```

```python
lengths = [len(line) for line in seq]
lengths
```

```python
plt.hist(lengths, bins=50)
```

```python
seq = pad_sequences(seq, maxlen=12)
seq.shape
```

```python
x = seq[:-1]
y = seq[1:]
```

```python
x.shape
```

```python
x = x.reshape(3415, 12, 1)
y = y.reshape(3415, 12, 1)
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
y_oh = to_categorical(y)
y[0]
```

```python
y_oh.shape
```

```python
model = Sequential()
model.add(LSTM(128, input_shape=(12, 1), return_sequences=True))
model.add(LSTM(256, return_sequences=True))
model.add(Dense(2801, activation='softmax'))    
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', metrics=['accuracy'], optimizer='adam')
```

```python
model.fit(x, y_oh, epochs=20, batch_size=20, validation_split=0.2 )
```

```python
pred = model.predict(x[7:8])
pred
```

```python
pred.shape
```

```python
pred_tok = np.argmax(pred, axis=2)
```
