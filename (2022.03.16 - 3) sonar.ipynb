```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import tensorflow as tf

from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('./Data/sonar.csv', header = None)
df.head()
```

```python
df.info()
```

```python
df.iloc[:, 60].value_counts()
```

```python
x = df.values[:, :60]
y = df.values[:, 60]
```

```python
y
```

```python
y.shape
```

```python
# binary
```

```python
enc = LabelEncoder()
y_enc = enc.fit_transform(y)
y_enc
```

```python
model = Sequential()
model.add(Dense(24, input_dim = 60, activation = 'relu'))
model.add(Dense(12, activation = 'relu'))
model.add(Dense(6, activation = 'relu'))
model.add(Dense(1,activation = 'sigmoid'))
model.summary()
```

```python
model.compile(loss = 'binary_crossentropy', metrics = ['accuracy'], optimizer = 'adam')
```

```python
model.fit(x.astype(float), y_enc, epochs = 10, batch_size = 10)
```

```python
# multiclassification
```

```python
y_oh = tf.keras.utils.to_categorical(y_enc)
y_oh
```

```python
y_oh.shape
```

```python
model = Sequential()
model.add(Dense(24, input_dim = 60, activation = 'relu'))
model.add(Dense(12, activation = 'relu'))
model.add(Dense(6, activation = 'relu'))
model.add(Dense(2, activation = 'softmax'))
model.summary()
```

```python
model.compile(loss = 'categorical_crossentropy', metrics = ['accuracy'], optimizer = 'adam')
```

```python
model.fit(x.astype(float), y_oh, epochs = 100, batch_size = 10)
```

```python
model.evaluate(x.astype(float), y_oh)
```

```python
# train_test_split
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x.astype(float), y_oh)
```

```python
model.fit(x_train, y_train, epochs = 100, batch_size = 10)
```

```python
model.evaluate(x_test, y_test)
```
