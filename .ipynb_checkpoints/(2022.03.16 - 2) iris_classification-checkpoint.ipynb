```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import tensorflow as tf

from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('./Data/iris.csv', names = ['sl', 'sw', 'pl', 'pw', 'species'])
df.head()
```

```python
sns.pairplot(df, hue = 'species')
```

```python
df.species.value_counts()        # label , multi classification
```

```python
y = df.values[:, 4]                         # values  array   .
x = df.values[:, :4]
```

```python
e = LabelEncoder()          ""       # y   
y_en = e.fit_transform(y)
y_en
```

```python
y_oh = tf.keras.utils.to_categorical(y_en)       # one-hot
y_oh
```

```python
y_oh.shape
```

```python
model = Sequential()
model.add(Dense(16, input_dim = 4, activation = 'relu'))     # Dense : fully connected layer
model.add(Dense(8, activation='relu'))
model.add(Dense(3, activation = 'softmax'))
model.summary()

# softmax
# https://blog.naver.com/chocola_meilleure/222220570152
```

```python
model.compile(loss = 'categorical_crossentropy', metrics = ['accuracy'], optimizer = 'adam')
```

```python
model.fit(x.astype(float), y_oh, epochs=100, batch_size = 10)       # numpy array => float  
```

```python
model.evaluate(x.astype(float), y_oh)
```
