```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from tensorflow.keras.layers import *
from tensorflow.keras.models import *
from tensorflow.keras.utils import *
from sklearn.preprocessing import *
```

```python
df = pd.read_csv('eighthr.data', header=None)
df.head()
```

```python
# neural network를 구성해서 예측하는 모델, LSTM으로 예측하는 모델을 각각 생성하고 평가하시오...
# target : 73번 칼럼
# object => numeric
# scaling
```

```python
df.head()
```

```python
df.info()
```

```python
df[1].apply(lambda x: 0 if x == '?' else x).unique()
```

```python
for col in df.columns[1:]:
  df[col] = df[col].apply(lambda x: 0 if x == '?' else x)
  df[col] = df[col].astype('float')
```

```python
df[0] = df[0].astype('datetime64')
```

```python
df.loc[:2000, 1:72]
```

```python
x_train, x_test = df.loc[:1999, 1:72], df.loc[2000:, 1:72]
```

```python
y_train, y_test = df.loc[:1999, 73], df.loc[2000:, 73]
```

```python
scaler = MinMaxScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.fit_transform(x_test)
```

```python
x_train_scaled = np.array(x_train_scaled)
x_test_scaled = np.array(x_test_scaled)
```

```python
x_train_scaled.reshape(2000, 72, 1)
```

```python
model = Sequential()
model.add(LSTM(128, input_shape = (72, 1)))
model.add(Dropout(0.2))
model.add(Dense(1, activation='sigmoid'))
model.summary()
```

```python
model.compile(loss='binary_crossentropy', metrics=['accuracy'], optimizer='adam')
```

```python
model.fit(x_train_scaled, y_train, epochs=20, batch_size=10, validation_split=0.2)
```
